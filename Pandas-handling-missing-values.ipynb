# isnull

```python
# iska use karke hum pata kar sakte hai ki hamare data set me kitni null values and kitni missing values hai 
```

```python
df = pd.read_csv('/kaggle/input/fortune-data-set/Fortune_10.csv')
df
```

```python
df.isnull()

# jaha pe true print hoga to uska matlab hai ki us data set me NaN values hai 
```

```python
# to find which column have the null values 
df.isnull().sum()

# isme null value hamara expenses wale column me hai 
```

```python
# to find total null values 
df.isnull().sum().sum()

# here one values is  null
```

# notnull

```python
# not nulll means totally oposite values 
# not null valuues present in the data set 
df = pd.read_csv('/kaggle/input/fortune-data-set/Fortune_10.csv')
df
```

```python
df.notnull()

# isme jaha par humko false dekhayega mens whaha ki value null hai 
```

```python
df.notnull().sum().sum()

# 79 values hai jo null nahi hai 

# this function is use for the data frame and the same function is also use for the series 
```

# series isnull and notnull

```python
# series
import numpy as np

sr = pd.Series([1,2,3,np.nan,4,5,np.nan,6])
sr
```

```python
sr.isnull()
```

```python
sr.notnull()
```

```python
# same as it is hum sum function ka bhi use kar sakte hai 

sr.isnull().sum().sum()
sr.notnull().sum().sum()
```

# dropna

```python
# data frame me jitni bhi messing values ya null values hoti hai us particular column or row ko drop kardeta hai 
# dropna

# reaad csv files
df = pd.read_csv('/kaggle/input/fortune-data-set/Fortune_10.csv')
df
 
```

```python
df.dropna()

# 9th row is deroped 
```

```python
# Drop_na ka first parameter hota hai wo axis hota hai 
# axis parameter me 0 ya 1 dete hai 1 stands for column and 0 stands for row 

df.dropna(axis = 1)

# jaise hamare expences ke column me nan value thi to us column ko isne remove kar diya 
```

```python
df.dropna(how = 'any') # apply on row

# means hamare data me jo bhi nan value hoga wo pura row drop ho jayega 
```

```python
# apply on column 

df.dropna(axis = 1,how = 'any')

# to isko apply karne se hamara column drop ho jayega 
```

```python
# row ke nan value ko drop karne ke liye hum df.dropna(subset) ka use karte hai

df.dropna(subset = ['ID'])
```

```python
df.dropna(subset = ['Expenses']) # 9th row will be deleted 
```
